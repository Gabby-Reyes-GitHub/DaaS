```python
df = pd.DataFrame()
titles = []
news = []

for n in noticias_clean:
    req = requests.get(n)
    res = BeautifulSoup(req.text)
    
    title = res.find('h1', {'class': 'jeg_post_title'})
    titles.append(title.contents)
    
    noticia = res.find_all('p', text=True)
    news.append(noticia)
    
df['URLs'] = noticias_clean
df['Titulos'] = titles
df['Contenidos'] = news  

df
```
